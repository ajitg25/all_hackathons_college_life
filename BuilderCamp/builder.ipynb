```python
d = ["Bank Charges",
"CARRIAGE",
"DOTCOM POSTAGE",
"Manual",
"PADS TO MATCH ALL CUSHIONS",
"POSTAGE"]
chrg_dict = dict()
def removing(x):
    for i in d:
        str_r = i.lower()
        print(str_r)
        if str_r in x:
            print(str_r) 
            if str_r in chrg_dict:
                
                chrg_dict[str_r] = 0 
            else:
                chrg_dict[str_r] +=1
            x = x.replace(str_r,'')
    return x 
```

```python
data = data[~((data.Quantity <0)  |(data.UnitPrice<=0) | (data.CustomerID.isnull()))]
```

```python
data_g = data.groupby(['CustomerID']).agg(no_bills=('InvoiceNo', pd.Series.nunique), 
                                 sum_bill=('bill_amount', np.sum),
                                Recency=('InvoiceDate', max),
                                 Recency_min=('InvoiceDate', min)).reset_index()
```

```python
data_g['Average_bill'] = data_g.sum_bill/data_g.no_bills
```

```python
data_g.loc[data_g.Recency.dt.year <2011,'Recency_int'] = 0
data_g.loc[data_g.Recency.dt.year >=2011,'Recency_int'] = data_g.Recency.dt.month
```

```python
data_g['month_interval'] = (data_g['Recency'] - data_g['Recency_min']).dt.days/30
```

```python
data_g['no_of_visit_per_month'] = data_g['month_interval']/data_g.
```
